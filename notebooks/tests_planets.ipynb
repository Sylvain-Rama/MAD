```python
from mad.objs.planets import Planet, PlanetConfig
from mad.configs.planets import EARTH_SETTINGS
```

```python
EARTH_SETTINGS["position"] = [0, 0] # Making it 2D
earth = Planet(PlanetConfig(**EARTH_SETTINGS))

earth
```

```python
launchsite = earth.create_2D_point(name="Launchpad", altitude=10)
target = earth.create_2D_point_at_distance(launchsite, distance_km=5000, name="Target")

launchsite, target, f"{earth.surface_distance(launchsite, target)=:.2f} m"
```

```python
fig = earth.plot_2D_with_points(points = [launchsite, target])
```
