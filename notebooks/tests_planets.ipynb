```python
from mad.objs.planets import Planet, PlanetConfig
from mad.objs.constants import EARTH_SETTINGS
```

```python
EARTH_SETTINGS["position"] = [0, 0] # Making it 2D
earth = Planet(PlanetConfig(**EARTH_SETTINGS))

earth, f"{earth.escape_velocity=:.2f} m/s", f"{earth.gravity_at_surface=:.2f} m/s^2"
```

```python
launchsite = earth.create_random_point(name="Launchpad", altitude=10)
target = earth.create_random_point_at_distance(launchsite, distance=7000_000, name="Target")

launchsite, target, f"{earth.surface_distance(launchsite, target)=:.2f} m"
```

```python
fig = earth.plot_2D_with_points(points = [launchsite, target])
```
